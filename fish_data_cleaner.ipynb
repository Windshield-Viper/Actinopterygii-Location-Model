```python
import pandas as pd
```

```python
# data has not been added to this repo as it's distributed by IUCN
fish_pts_df = pd.read_csv("/content/drive/MyDrive/Data/FW_FISH_points.csv")
```

```python
less_rows_fish = fish_pts_df.drop(columns=["objectid", "id_no", "compiler", "yrcompiled", "citation", "subspecies", "subpop", "source", "basisofrec", "catalog_no", "dist_comm", "island", "tax_comm", "legend", "kingdom", "phylum", "class"])
```

```python
clean_fish_data = less_rows_fish.dropna(axis="rows", subset=["event_year"])
```

```python
clean_fish_data.describe()
```

```python
clean_fish_data.head()
```

```python
clean_fish_data.reset_index(drop=True, inplace=True)
#del clean_fish_data["level_0"]
```

```python
clean_fish_data.head()
```

```python
#clean_fish_data.to_csv("clean_fish_data.csv")
```

```python
clean_fish_data.rename(columns={'order_': 'order'}, inplace=True)
```

```python
clean_fish_data.head()
```

```python
clean_fish_data.describe(include=[object])
```

```python
clean_fish_data.drop(columns=["sci_name"], inplace=True)
```

```python
clean_fish_data = clean_fish_data[clean_fish_data.event_year != 0]
```

```python
clean_fish_data.head()
```

```python
clean_fish_data['order_as_num'] = pd.factorize(clean_fish_data['order'])[0]
clean_fish_data['family_as_num'] = pd.factorize(clean_fish_data['family'])[0]
clean_fish_data['genus_as_num'] = pd.factorize(clean_fish_data['genus'])[0]
clean_fish_data['category_as_num'] = pd.factorize(clean_fish_data['category'])[0]
```

```python
clean_fish_data.drop(columns=["order", "family", "genus", "category"], inplace=True)
```
